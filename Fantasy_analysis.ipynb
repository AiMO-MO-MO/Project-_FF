### Visualization

```python
# Create bar graph from "Fullcombine_df" dataframe for players' position and 40 yard touchdown

position_40yd = FullCombine_df.groupby(["Pos"])["40yd"].count().reset_index().sort_values(by = "40yd")
position_40yd
```

```python
plt.figure(figsize=(10,5))
plt.bar(position_40yd["Pos"], position_40yd["40yd"])
plt.xlabel("Position of player")
plt.ylabel("40 yard touchdown")
plt.title("Player position Vs. Touchdown")
```

```python
# Convert Height column to float from FullCombine DataFrame

FullCombine_df["Ht"].replace("-",".",regex = True,inplace=True)
FullCombine_df = FullCombine_df.astype({"Ht":float})
FullCombine_df.info()
```

```python
height_position_40yd = FullCombine_df.groupby(["Pos","Ht"])["40yd"].count().reset_index().sort_values(by = "Ht")
height_position_40yd
```

```python
plt.figure(figsize=(10,5))
#plt.bar(height_position_40yd["Pos"], height_position_40yd["Ht"])
plt.scatter(height_position_40yd["Ht"], height_position_40yd["40yd"])
plt.xlabel("Height of player")
plt.ylabel("Number of Touchdown")
plt.title("Player height Vs. Touchdown")
#plt.ylim(6,7)
plt.xlim(4.5,7.5)
plt.show()
```
