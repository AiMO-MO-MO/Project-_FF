```python
FortyTime = WR['40yd']
FantasyPoints = WR['Total Fantasy Points']
plt.title("WR Fantasy Production by 40 Time")
(slope,intercept,rvalue,pvalue,stderr) = linregress(FortyTime,FantasyPoints)
regress_values = FortyTime * slope + intercept
plt.plot(FortyTime,regress_values,"r-")
line_eq = "y=" + str(round(slope,2))+"x + " + str(round(intercept,2))
rsquared1 = "r-squared = " + str(round((rvalue ** 2),3))
plt.annotate(line_eq, (4.2,3500), fontsize = 15, color = "red")
plt.annotate(rsquared1, (4.2,3000), fontsize = 15, color = "red")
plt.scatter(FortyTime,FantasyPoints)
plt.xlim(4.2,4.8)
plt.xlabel("Forty Time")
plt.ylabel("Career Fantasy Points")
plt.show()
```
