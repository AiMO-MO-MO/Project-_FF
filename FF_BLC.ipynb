```python
import pandas as pd
import matplotlib.pyplot as plt
import os 
import csv
import numpy as np
import seaborn as sns
```

```python
FullCombine_df = pd.read_csv("DataFrames/FullCombine.csv")
del FullCombine_df["Unnamed: 0"]
FullCombine_df.set_index("Player")
```

```python
FullFFMerge = pd.read_csv("DataFrames/FullFFMerge.csv")
del FullFFMerge["Unnamed: 0"]
FullFFMerge
```

```python
# Created a smaller dataframe sample years (2000,2009) and perspective Fantasy Points

reducedFF_df = FullFFMerge[["Player","Pos","Age_2000","FantasyPoints_2000","Age_2009", 
                            "FantasyPoints_2009", "Total Fantasy Points",]]
reducedFF_df.set_index("Player")
reducedFF_df.head(24)
```

```python
reducedFF_df.groupby(["Player","Pos"])["Total Fantasy Points"]
reducedFF_df.head(36)
```

```python
# Sorting Fantasy Points in ascending order starting at yr 2000 to yr 2009

reducedFF_df.sort_values(by=["FantasyPoints_2000","FantasyPoints_2009"], ascending=[False,False]).set_index("Player")
```

```python
reducedFF_df.plot(kind="scatter", x="Age_2000", y="Total Fantasy Points", marker="^", s= 8, c= "red", 
                  grid=False, figsize=(6,6), title= "Age vs. Total Points")
plt.xlim(20,40)
plt.xlabel("Players Age in 2000")
plt.show()
```

```python
reducedFF_df.plot(kind="scatter", x="Age_2009", y="Total Fantasy Points", marker= "d", s= 11, c= "blue", 
                  grid=False, figsize=(6,6), title= "Age vs. Total Points")
plt.xlim(20.5,40.5)
plt.xlabel("Players Age in 2009")
plt.show()
```

```python
selected_positions = ["QB", "WR", "K", "TE", "RB"]
filtered_df = FullCombine_df[FullCombine_df["Pos"].isin(selected_positions)]

filtered_df_sorted = filtered_df.sort_values(by="Pos")

x_axis = filtered_df_sorted["Pos"]
y_axis = filtered_df_sorted["40yd"]

plt.scatter(x_axis, y_axis, marker="d", facecolors="red", alpha= .5)
plt.ylim(4,6.5)
plt.title("Position vs 40 yard")
plt.xlabel("Position")
plt.ylabel("40 Yard")
plt.show()
```

```python
FullCombine_df_sorted = FullCombine_df.sort_values(by="Pos")
x_axis = FullCombine_df_sorted["Pos"]
y_axis = FullCombine_df_sorted["40yd"]
plt.scatter(x_axis, y_axis, marker="d", facecolors="red", alpha= .5)
plt.ylim(4,6.5)
plt.title("Position vs 40yd")
plt.xlabel("Position")
plt.ylabel("40 Yard")
plt.show()
```

```python
selected_positions = ["QB", "WR", "TE", "RB"]
filtered_df = reducedFF_df[reducedFF_df["Pos"].isin(selected_positions)]

position_dist = reducedFF_df["Pos"].value_counts()

labels = position_dist.index  # Use the index of the Series for labels
positions = position_dist.values
colors = plt.cm.rainbow(range(len(position_dist)))  # Use a colormap for colors
explode = (0.1,) * len(position_dist)  # Adjust the explode values

position_count = plt.pie(positions, labels=labels, colors=colors, explode=explode,
                         autopct="%1.1f%%", shadow=True, startangle=150)

plt.legend(loc="lower right")
plt.title("Position Distribution")

plt.show()
```
