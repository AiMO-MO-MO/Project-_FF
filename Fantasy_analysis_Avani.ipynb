```python
# Top 10 players who scored higest Fantasy Points

fantasy_point = FullFFMerge.groupby(["Player"])["Total Fantasy Points"].sum().sort_values(ascending=False).reset_index()
fantasy_point.head(10)
```

```python
# Find Fantasy points for each player for year 2020,2019,2018 and 2017

def calculate_fantasy_points(df, year, column_name):
    fantasy_points = df.groupby(["Player", "Pos"])[column_name].sum().sort_values(ascending=False).reset_index()
    fantasy_points = fantasy_points.loc[fantasy_points[column_name] != 0]
    return fantasy_points

# Create dataframes
fantasy_point_2020 = calculate_fantasy_points(FullFFMerge, "FantasyPoints_2020", "FantasyPoints_2020")
fantasy_point_2019 = calculate_fantasy_points(FullFFMerge, "FantasyPoints_2019", "FantasyPoints_2019")
fantasy_point_2018 = calculate_fantasy_points(FullFFMerge, "FantasyPoints_2018", "FantasyPoints_2018")
fantasy_point_2017 = calculate_fantasy_points(FullFFMerge, "FantasyPoints_2017", "FantasyPoints_2017")
#fantasy_point_2017
```

```python
# Add suffixes to identify the origin of columns
suffixes_list = ['_2020', '_2019', '_2018', '_2017']
df_list = [fantasy_point_2020, fantasy_point_2019, fantasy_point_2018, fantasy_point_2017]

# Use a loop to merge DataFrames with suffixes
merged_2020_19_18_17 = df_list[0]
for i in range(1, len(df_list)):
    merged_2020_19_18_17 = pd.merge(merged_2020_19_18_17, df_list[i], on="Player", how="inner", suffixes=('', suffixes_list[i]))
  
merged_2020_19_18_17
```

```python
# Merge all four years fantasy points dataframe to find consistant players who are getting higher score

merged_2020_19_18_17 = merged_2020_19_18_17.drop(columns=['Pos_2019','Pos_2018','Pos_2017'])
merged_2020_19_18_17
```

```python
# Create a box plot for Fantasy Points for years 2020,2019,2018,2017 for players who played all 4 years

merged_2020_19_18_17.plot.box(x = "Player",figsize = (10,5))
plt.ylabel("Total Fantasy Point")
plt.title("Fantasy Points for years 2020, 2019, 2018, 2017")
plt.show()
```

```python
# Create subplots for Fantasy Points Vs. Players 

merged_2020_19_18_17.plot(x = "Player",subplots = True,figsize = (15,7),\
                          ylabel="Fantasy Points",title = "Fantasy Points Vs. Players")
```

```python
# Find mean Fantasy score for each player for 2020,2019,2018,2017

merged_2020_19_18_17["Total_FantasyPoints"] = merged_2020_19_18_17[["FantasyPoints_2020","FantasyPoints_2019","FantasyPoints_2018","FantasyPoints_2017"]].sum(axis=1)
merged_2020_19_18_17["Average_FantasyPoints"] = merged_2020_19_18_17[["FantasyPoints_2020","FantasyPoints_2019","FantasyPoints_2018","FantasyPoints_2017"]].mean(axis=1).sort_values()
merged_2020_19_18_17
```

```python
# Create a bar graph which shows top 10 players' Fantasy Points

top_10_players = merged_2020_19_18_17.drop("Total_FantasyPoints",axis = 1).head(10)
#top_10_players
top_10_players.plot.bar(x = "Player",figsize = (10,5))
plt.ylabel("Fantasy Points")
plt.title("Top 10 player's Fantasy Points for 2020,2019,2018,2017")
plt.show()
```

```python
# Find Total Fantasy Points for years 2011-2020

def calculate_sum_of_fantasy_points(df):
    sum_2011 = df["FantasyPoints_2011"].sum()
    sum_2012 = df["FantasyPoints_2012"].sum()
    sum_2013 = df["FantasyPoints_2013"].sum()
    sum_2014 = df["FantasyPoints_2014"].sum()
    sum_2015 = df["FantasyPoints_2015"].sum()
    sum_2016 = df["FantasyPoints_2016"].sum()
    sum_2017 = df["FantasyPoints_2017"].sum()
    sum_2018 = df["FantasyPoints_2018"].sum()
    sum_2019 = df["FantasyPoints_2019"].sum()
    sum_2020 = df["FantasyPoints_2020"].sum()

    sum_years = pd.DataFrame({
        "Years": ["2011", "2012", "2013", "2014", "2015", "2016", "2017", "2018", "2019", "2020"],
        "Fantasy_Points": [sum_2011, sum_2012, sum_2013, sum_2014, sum_2015, sum_2016, sum_2017, sum_2018, sum_2019, sum_2020]
    })

    return sum_years

# Example usage:
sum_years_total = calculate_sum_of_fantasy_points(FullFFMerge)
sum_years_total
```

```python
# Create a line plot for Total Fantasy Points from 2011-2020

x = sum_years_total["Years"]
y = sum_years_total["Fantasy_Points"]

plt.figure(figsize=(10,5))
plt.plot(x,y, marker = "o")
plt.xlabel("Years")
plt.ylabel("Total Fantasy Points")
plt.title("Total Fantasy Points from 2011-2020")
plt.show()
```
